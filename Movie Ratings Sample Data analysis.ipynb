```python
# Analysis of a movie review database
# Use Movielense smaller file of 100k reviews of movies
# GroupLens Research has collected and made available rating data sets from the MovieLens (http://movielens.org).
# Small dataset: 100,000 ratings and 3,600 tag applications applied to 9,000 movies by 600 users. Last updated 9/2018.
# Contains 4 csv files with genre etc., ratings.csv, tags.csv, links.csv
# Also read users.csv (from a folder http://grouplens.org/datasets/movielens/, movies.csv)

import numpy as np
from pandas import Series
import pandas as pd
from pandas import DataFrame

#Read user data
# pass in column names for each CSV
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']

users = pd.read_csv(
    'http://files.grouplens.org/datasets/movielens/ml-100k/u.user', 
    sep='|', names=u_cols)

users.head()
```

```python
# Read the ratings data table from the file 
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv(
    'http://files.grouplens.org/datasets/movielens/ml-100k/u.data', 
    sep='\t', names=r_cols)

ratings.head()
```

```python
# Read the data about the movies
# the movies file contains columns indicating the movie's genres
# let's only load the first five columns of the file with usecols
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']

movies = pd.read_csv(
    'http://files.grouplens.org/datasets/movielens/ml-100k/u.item', sep ='|', na_values= None , names=m_cols, usecols=range(5))
    
movies.head()
```
